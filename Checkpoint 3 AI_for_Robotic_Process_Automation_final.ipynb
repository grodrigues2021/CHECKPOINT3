```python
headers = ['Month','Day', 'Artist', 'Album'] #CRIEI OS LABELS DO DATAFRAME
df = pd.DataFrame(columns = headers)# DEFININDO MINHA LISTA COMO TITULO DAS COLUNAS

coluna_num = 0 # INICIALIZANDO UMA VARIÁVEL QUE SERÁ UTILIZADA MAIS ABAIXO

for i in ['table_January','table_February','table_March','table_April','table_May','table_June','table_July','table_August','table_September','table_October']:
    
    tables = soup.find('table', {'id': i } )# aqui eu to pegando o m~es que será iterado utilizando a variavel i
    for row in tables.find_all('tr')[1:]:#pego todos os tr pulando a primeira que são os headers
        
        data = row.find_all('td')#pego todos os tds dentro da tr
        row_data = [td.text.strip() for td in data ]#monto a linha pegando todos os td e utilizaei o strip de recomendação para tirar espações antes e depois mas não sei se era tão util nesse caso.
        
        if len(row_data) == 3:# aqui reolvo o seguinproblema do rowspan,  se tiver 3 td é por que é onde esta o mÊs
            coluna_num=row_data[0]#peguei o valor do mês
        else:
            row_data.insert(0,coluna_num)#aqui estou inserindo o valor do mês na linha
                  
        lenght = len(df)#aqui eu pego o tamanho do datafrae
        
        row_data.insert(0,i[6:])#aqui eu inseri o nome do mês no dataframe
                       
        df.loc[lenght] = row_data#aqui eu monto o data frame onde eu vou pegando linha a linha e inserindo o row_data para montar o dataframe final
```

## VERIFICANDO A DATA PARA ENVIO DO EMAIL

No conceito geral eu usei o datetime  para pegar  o mês e o dia atual quando o programa executa e comparei ele com  os valores no data frame para enviar o email.

Tem um pequeno problmea aqui,  como cada album tem  a data do dia  de lançamento repetida no dataframe  a pessoa vai receber vários emails no mesmo dia.

Como eu falei anteriormente não t abonito, mas a pessoa será lembrada do lançamento.

Esse seria o próximo passo para resolver,  mandar um único email.


```python
month = (datetime.now().month) #pega o valor do mês em formato número
mes = calendar.month_name[month] # para a variavel anterior e transforma o mês em nome para comparar com os valores do dataframe.

data_anterior = 1 - datetime.now().day # define a data de lançamento para data -1


col_data = df.iloc[:,1]# peguei a coluna de dia de lançamento
col_mes = df.iloc[:,0] #peguei a coluna de mês de lançamento


for x,y in zip(col_mes,col_data):#iteri as duas colunas ao mesmo tempo
    data_lancamento = int(y)# a data estava vindo como string deu uma dor de cabeça , transformei em int
    
    if (x == mes) and (data_lancamento == data_anterior): # se o valor do mês atual e a data atual forem iguais ao mês e data do data frame ele envia o email executando a função.
        enviar_email()
```

## CRIANDO O ENVIO DE EMAIL


```python
def enviar_email():#criei uma função para envia ro email

    import smtplib, ssl

    port = 465  # For SSL
    smtp_server = "coloque o endereço do seu servidor smtp"
    sender_email = "coloque seu email de envio"  # Enter your address
    receiver_email = "coloque o email de quem vai receber"  # Enter receiver address
    password = "coloque sua senha"

    message = "Subject: AVISO DE LANCAMENTO NOVO AMANHA\n\n"
  
    message += "NÂO ESQUEÇA AMANHA SERÁ O LANCAMENTO DE UM NOVO ALBUM"
    
    context = ssl.create_default_context()
    with smtplib.SMTP_SSL(smtp_server, port, context=context) as server:
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message)
```
